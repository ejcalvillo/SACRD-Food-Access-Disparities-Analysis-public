```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# === Step 1: Load your dataset ===
df = pd.read_csv("../Dashboard_Final/avg_gap_scores.csv")  
df = df.dropna(subset=['interaction_count', 'share_count', 'population', 'program_count'])

# === Step 2: Select relevant columns for clustering ===
features = df[['interaction_count', 'share_count', 'population','program_count']]
df['share_count'] *= 10

# === Step 3: Scale the features ===
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# === Step 4: Choose number of clusters (try 3 for now) ===
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled_features)
sns.pairplot(df, vars=features, hue="cluster", palette="Set1")
plt.suptitle("KMeans Clustering of ZIP Codes", y=1.02)
plt.show()


# === Step 5: Optional - visualize clusters in 2D ===
# We'll use just the first two components for visualization
plt.figure(figsize=(8, 6))
plt.scatter(df['interaction_count'], df['share_count'], c=df['cluster'], cmap='viridis')
plt.xlabel('interaction_count')
plt.ylabel('share_count')
plt.title('Clustering Zip Codes')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()

```
