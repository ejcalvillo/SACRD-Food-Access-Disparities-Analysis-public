```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

df["zipcode"] = df["zipcode"].astype(str).str.strip()
df = df[["zipcode", "gap_score", "demand_score", "program_count", "population"]].dropna()

features = ["gap_score", "demand_score", "program_count", "population"]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[features])

inertia = []
K_range = range(1, 10)

for k in K_range:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(scaled_data)
    inertia.append(model.inertia_)
    
for k, val in zip(K_range, inertia):
    print(f"k={k}: inertia={val:.2f}")


plt.figure(figsize=(6, 4))
plt.plot(K_range, inertia, marker='o')
plt.title("Elbow Method for Optimal K")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()

```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df["cluster"] = kmeans.fit_predict(scaled_data)

sns.pairplot(df, vars=features, hue="cluster", palette="Set1")
plt.suptitle("KMeans Clustering of ZIP Codes", y=1.02)
plt.show()
```

```python
cluster_summary = df.groupby("cluster")[features].mean().round(2)
print(cluster_summary)
```

```python
zipcodes_by_cluster = df.groupby("cluster")["zipcode"].apply(list)
zipcodes_by_cluster
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(scaled_data, df["cluster"])
print("Silhouette Score:", score)
```
